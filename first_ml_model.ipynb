```python
#import libraries
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
melbourne_file_path="melb_data.csv"
melbourne_data=pd.read_csv(melbourne_file_path)
```

```python
melbourne_data.head()
```

```python
melbourne_data.describe()
```

```python
melbourne_data.info()
```

```python
#drop missing values
melbourne_data.dropna(axis=0,inplace=True)
```

```python
#print all columns
melbourne_data.columns
```

```python
#select predictin target
y=melbourne_data.Price
```

```python
#Add features
melbourne_features=['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X=melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```

# devide data into train and test

```python
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
```

# Building model

```python
melbourne_model=DecisionTreeRegressor(random_state=0)
melbourne_model.fit(train_X,train_y)
```

# Predict price

```python
val_prediction=melbourne_model.predict(val_X)
print(val_prediction)
```

# Calculate Mean Absolute Error

```python
mae=mean_absolute_error(val_y,val_prediction)
print(mae)
```
